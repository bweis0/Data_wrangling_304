## Initial Graph Issue ##

The initial issue seems to be that dates may be out of order and there was a large event count leading up to a point.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('system_of_record_timestamps.csv')

data['Parsed_Timestamp'] = pd.to_datetime(data['Primary_Timestamp'], errors='coerce')

data = data.dropna(subset=['Parsed_Timestamp'])

data['Event_Count'] = pd.to_numeric(data['Event_Count'], errors='coerce')
data = data.dropna(subset=['Event_Count'])

plt.figure(figsize=(10, 6))
plt.plot(data['Parsed_Timestamp'], data['Event_Count'], marker='o', linestyle='-', color='b')
plt.title("Event Count Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Event Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

## Fixed Graph ##

After spending a while on troubleshooting dtypes and learning about the timezone functions (tz), I used a lambda function to make the times uniform that way I could organize them by date and get a clean graph.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("system_of_record_timestamps.csv")

df['Parsed_Timestamp'] = pd.to_datetime(df['Primary_Timestamp'], errors='coerce')

df['Parsed_Timestamp'] = df['Parsed_Timestamp'].apply(
    lambda x: pd.Timestamp(x).tz_localize('UTC') if x.tzinfo is None else x
)

df = df.dropna(subset=['Parsed_Timestamp'])

df['Event_Count'] = pd.to_numeric(df['Event_Count'], errors= 'coerce')
df = df.dropna(subset=['Event_Count'])

df = df.sort_values(by='Parsed_Timestamp')

plt.figure(figsize=(10, 6))
plt.plot(df['Parsed_Timestamp'], df['Event_Count'], marker='o', linestyle='-', color='b')
plt.title("Event Count Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Event Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```
